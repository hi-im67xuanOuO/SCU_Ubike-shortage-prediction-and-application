```python
a = []
b = []
c = []
d = []
e = []
f = []
g = []

for i in range(len(df['sno'])):
    if df['station'][i] == '466920':
        a.append(df['sno'][i])
    elif df['station'][i] == 'C0A980':
        b.append(df['sno'][i])
    elif df['station'][i] == 'C0A9C0':
        c.append(df['sno'][i])
    elif df['station'][i] == 'C0A9E0':
        d.append(df['sno'][i])
    elif df['station'][i] == 'C0A9F0':
        e.append(df['sno'][i])
    elif df['station'][i] == 'C0AC70':
        f.append(df['sno'][i])
    elif df['station'][i] == 'C0AC80':
        g.append(df['sno'][i])
    else:
        print(df['sno'])
```

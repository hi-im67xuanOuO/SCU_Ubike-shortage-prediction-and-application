```python
import pandas as pd
```

```python
df = pd.read_csv("缺車率計算.csv")
df.head()
```

```python
sna_unique = list(df['sna'].unique())# unique sna
tot_ = []
sarea_ = []
lat_ = []
lng_ = []
ar_ = []

sna_unique
```

```python
len(sna_unique) ##總共車站數
```

```python
times = 1 ##執行次數

## 靜態欄位資訊
for i in sna_unique:
    try:
        tot_.append(list(df[df['sna'] == i]['tot'].unique())[0])
        sarea_.append(list(df[df['sna'] == i]['sarea'].unique())[0])
        lat_.append(list(df[df['sna'] == i]['lat'].unique())[0])
        lng_.append(list(df[df['sna'] == i]['lng'].unique())[0])
        ar_.append(list(df[df['sna'] == i]['ar'].unique())[0])
    except:
        print(i)
    
    print(times)
    times+=1
```

```python
df_new = pd.DataFrame({"sna":sna_unique,'sarea':sarea_,'lat':lat_,'lng':lng_,'ar':ar_})
df_new
```

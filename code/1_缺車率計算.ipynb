```python
import pandas as pd
```

```python
df = pd.read_csv("bike_WeekdayHourMin.csv")
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
mydict = df.to_dict('records') ## 轉換格式處理效能較佳（好很多唷）
```

```python
mydict[0]
```

```python
for i in range(len(mydict)):
    mydict[i]['sbi_tot'] = mydict[i]['sbi']/mydict[i]['tot']
    mydict[i]['bemp_tot'] = mydict[i]['bemp']/mydict[i]['tot']

mydict[0]
```

```python
mydict
```

```python
## 轉回df格式
df_new = pd.DataFrame.from_dict(mydict)
df_new
```

```python
df_new.to_csv("缺車率計算.csv")
```
